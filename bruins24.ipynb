```python
from selenium import webdriver
import pandas as pd

url = "https://www.hockey-reference.com/teams/BOS/2024.html"

# Use a headless browser (Chrome in this example)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)
driver.get(url)

# Wait for the page to load (you may need to adjust the sleep duration)
import time
time.sleep(10)

# Extract tables using pandas read_html
tables = pd.read_html(driver.page_source)

# Close the browser
driver.quit()

# Filter for the "Skater Advanced Statistics" table
skater_advanced_df = [table for table in tables if 'Skater Advanced Statistics' in str(table)]

# Now you can access the DataFrame for analysis
if skater_advanced_df:
    skater_advanced_df = skater_advanced_df[0]
    print("Skater Advanced Statistics Table:")
    print(skater_advanced_df)
else:
    print("Skater Advanced Statistics table not found.")
```
